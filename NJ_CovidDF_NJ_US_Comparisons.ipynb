```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
Cleaned_Covid_Data = Path('Cleaned_Covid_Data.csv')
```

```python
Cleaned_Covid_Data
```

```python
Cleaned_Covid_Data = pd.read_csv(Cleaned_Covid_Data)
```

```python
Cleaned_Covid_Data.head()
```

```python
Age_Tot_Deaths2 = Cleaned_Covid_Data[["Age Group", "State", "Total Deaths", "COVID-19 Deaths", "Year", "Start Date", "End Date", "Sex"]] 
```

```python
Age_Tot_Deaths2
```

```python
columns = ["Age Group", "State", "Total Deaths", "COVID-19 Deaths", "Year", "Start Date", "End Date", "Sex"]
US = Age_Tot_Deaths2.loc[Age_Tot_Deaths2["State"] == "United States", columns]
US
```

```python
columns = ["Age Group", "State", "Total Deaths", "COVID-19 Deaths", "Year", "Start Date", "End Date", "Sex"]
NJ_df = Age_Tot_Deaths2.loc[Age_Tot_Deaths2["State"] == "New Jersey", columns]
NJ_df
```

```python
US_df_sd = US.loc[US["Start Date"] == "01/01/2020", columns]
US_df_sd
```

```python
NJ_df_sd = NJ_df.loc[NJ_df["Start Date"] == "01/01/2020", columns]
NJ_df_sd
```

```python
NJ_df_ed = NJ_df_sd.loc[NJ_df_sd["End Date"] == "09/23/2023", columns]
NJ_df_ed
```

```python
US_df_ed = US_df_sd.loc[US_df_sd["End Date"] == "09/23/2023", columns]
US_df_ed
```

```python
NJ_df_full = NJ_df_ed.loc[NJ_df_sd["Sex"] == "All Sexes", columns]
NJ_df_full
```

```python
US_df_full = US_df_ed.loc[US_df_ed["Sex"] == "All Sexes", columns]
US_df_full
```

```python
US_ages = US_df_full.drop([0, 1, 2, 3, 4, 6, 8, 10, 12]) 
US_ages
```

```python
total_deaths_only = NJ_df_full.drop([1150, 1151, 1152, 1153, 1154, 1155, 1156, 1157, 1158, 1159, 1160, 1161], axis=0)
total_deaths_only
```

```python
NJ_clean_ages = NJ_df_full.drop([1149, 1151, 1153, 1155, 1157], axis=0)
NJ_clean_ages
```

```python
total_deaths_only = ["Total Deaths", "COVID-19 Deaths"] 
total_deaths_only = [312963, 35691]
labels = ["New Jersey Total Deaths (2020-2023)", "New Jersey COVID-19 Deaths (2020-2023)"]
colors = ["yellow","green"]
explode = (0,0)
legend = 'Total Deaths vs. Covid Deaths'
```

```python
plt.pie(total_deaths_only, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)

plt.axis("equal")
plt.show()
```

```python
#Total death rates overall and related to COVID-17 in the US and NJ overall are relatively the same. 
#In comparison there is a 7% difference in COVID deaths where NJ trends slightly higher.
```

```python
NJ_clean_ages.plot(x="Age Group", y=["Total Deaths", "COVID-19 Deaths"],
                xlabel="NJ Age Group",
                ylabel="NJ Number of Deaths", kind="bar", stacked=True, 
                title="New Jersey Total Deaths vs. New Jersey Covid Deaths (2020-2023)",
                width = .7, colormap='Paired')

US_ages.plot(x="Age Group", y=["Total Deaths", "COVID-19 Deaths"],
            xlabel="US Age Group",
            ylabel="US Number of Deaths", kind="bar", stacked=True,
            title="United States Total Deaths vs. United States Covid Deaths (2020-2023)",
            width = .7, colormap="viridis")

plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

plt.show()
```

```python
#When comparing death rates in the US and NJ the trend is similar across ages as well. With death rates rising the higher the age. 
```

```python
US_ages.plot(x="Age Group", y=["Total Deaths", "COVID-19 Deaths"],
                xlabel="US Age Group",
                ylabel="US Number of Deaths",
                title="United States Total Deaths vs. United States Covid Deaths (2020-2023)",
                figsize=(15,5))

plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

NJ_clean_ages.plot(x="Age Group", y=["Total Deaths", "COVID-19 Deaths"],
                   xlabel="NJ Age Group",
                   ylabel="NJ Number of Deaths", kind="line",
                   title="New Jersey Total Deaths vs. New Jersey Covid Deaths (2020-2023)",
                   figsize=(15,5))
plt.show()
```

```python
#This chart shows how the rate and trend of rising death rates in the US overall is very similar to the death rates both 
#overall and with COVID in NJ. 
#Minor differences arise with New Jersey having a higher rate of deaths among the 85 and over group.
```
