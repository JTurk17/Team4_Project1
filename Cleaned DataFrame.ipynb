```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
covid_data_csv = Path("COVID-19_Deaths_by_Sex_and_Age_20240328.csv")
covid_data_df = pd.read_csv(covid_data_csv)
```
