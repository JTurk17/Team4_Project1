```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
covid_data_csv = Path("Cleaned_Covid_Data.csv")
covid_data_df = pd.read_csv(covid_data_csv)
```

```python
covid_data_df.head()
```

```python
columns = ["Data As Of", "Start Date", "End Date",	"Group", "Year", "Month", "State", "Sex", "Age Group",
           "COVID-19 Deaths", "Total Deaths", "Pneumonia Deaths", "Pneumonia and COVID-19 Deaths"]
covid_nj_df = covid_data_df.loc[covid_data_df["State"] == "New Jersey", columns]
```

```python
covid_nj_df
```

```python
columns2 = ["Start Date", "End Date", "State", "Sex","Group", "Age Group", "Year", "Month", "COVID-19 Deaths",
            "Pneumonia Deaths", "Pneumonia and COVID-19 Deaths"]
#creating data fram for males
male_df = covid_nj_df.loc[covid_nj_df["Sex"] == "Male", columns2]
male_df
```

```python
female_df = covid_nj_df.loc[covid_nj_df["Sex"] == "Female", columns2]
female_df
```

```python
total_mcovid_deaths = pd.DataFrame(male_df.loc[male_df['Age Group'] == "All Ages",:])
total_mcovid_deaths.reset_index(inplace = True)

total_mcovid_deaths.head()
```

```python
by_year_mdeaths = total_mcovid_deaths.loc[total_mcovid_deaths["Group"] == "By Year", columns2]
by_year_mdeaths ['Year'] = by_year_mdeaths ['Year'].astype(int)

by_year_mdeaths
```

```python
c = ['maroon', 'olive','lightsteelblue']
covid_male_deaths = by_year_mdeaths.plot.bar(x='Year', y=['Pneumonia and COVID-19 Deaths','Pneumonia Deaths','COVID-19 Deaths'],
                                             figsize=(10,6), alpha=.8, color=c,title ='Male Deaths in NJ')
plt.xlabel('Year')
plt.ylabel('# of Deaths')
plt.savefig("Male Deaths.png")
```

```python
total_fcovid_deaths = pd.DataFrame(female_df.loc[female_df['Age Group'] == "All Ages",:])
total_fcovid_deaths.reset_index(inplace = True)

total_fcovid_deaths
```

```python
by_year_fdeaths = total_fcovid_deaths.loc[total_fcovid_deaths["Group"] == "By Year", columns2]
by_year_fdeaths ['Year'] = by_year_fdeaths ['Year'].astype(int)

by_year_fdeaths
```

```python
c2 = {'orange','red', 'green'} 
by_year_fdeaths.plot.bar(x='Year', y=['Pneumonia and COVID-19 Deaths','Pneumonia Deaths','COVID-19 Deaths'],
                                             figsize=(10,6), alpha=.8, color=c2, title ='Female Deaths in NJ')
plt.xlabel('Year')
plt.ylabel('# of Deaths')
plt.ylim(0,10000)
plt.savefig("Female Deaths.png")
```

```python
all_deaths = pd.concat([by_year_fdeaths,by_year_mdeaths])
#ax = all_deaths.plot(kind='bar', x='Year', y=['Pneumonia and COVID-19 Deaths','Pneumonia Deaths','COVID-19 Deaths'])
all_deaths
```

```python
by_year_fdeaths['All Deaths']= by_year_fdeaths[['COVID-19 Deaths', 'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths']].sum(axis=1)
by_year_fdeaths
```

```python
all_fdeaths = by_year_fdeaths['All Deaths'].sum()
all_fdeaths
```

```python
by_year_mdeaths['All Deaths']= by_year_mdeaths[['COVID-19 Deaths', 'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths']].sum(axis=1)
by_year_mdeaths
```

```python
all_mdeaths = by_year_mdeaths['All Deaths'].sum()
all_mdeaths
```

```python
deaths = {'Sex': ['Male','Female'], 'Deaths': [all_mdeaths,all_fdeaths]}
all_fmdeaths = pd.DataFrame(deaths)
all_fmdeaths = all_fmdeaths.set_index(['Sex'])
all_fmdeaths ['Deaths'] = all_fmdeaths ['Deaths'].astype(int)
all_fmdeaths
```

```python
colors3= ['lightsteelblue','orange']
all_fmdeaths.plot(kind='pie', autopct='%1.1f%%', y='Deaths', title='Combined Deaths from 2020-2023 by Gender',
                  subplots=True, colors=colors3,shadow=True, legend='', label='')
plt.savefig("Percentage of Deaths.png")
```
