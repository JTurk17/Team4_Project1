```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
covid_data_csv = Path("Provisional_COVID-19_Deaths_by_Sex_and_Age_20240328.csv")
covid_data_df = pd.read_csv(covid_data_csv)
```

```python
covid_data_df.head(20)
```

```python
covid_data_df.columns
```

```python
new_df = covid_data_df.loc[:,['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month',
       'State', 'Sex', 'Age Group', 'COVID-19 Deaths', 'Total Deaths',
       'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths']]
new_df
```

```python
new_df['Year'] = new_df['Year'].fillna(0)
new_df['Month'] = new_df['Month'].fillna(0)
new_df
```

```python
new_df.columns
```

```python
new_df.dropna(subset = ['COVID-19 Deaths', 'Total Deaths',
       'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths'], inplace=True)
new_df
```

```python
new_df.to_csv("Cleaned_Covid_Data.csv")
```
