```python
import pandas as pd
import numpy as np 
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
covid_data_csv = Path("Cleaned_Covid_Data.csv")
covid_data_df = pd.read_csv(covid_data_csv)
```

```python
covid_data_df
```

```python
final_data_df = covid_data_df.loc[:,['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month',
       'State', 'Sex', 'Age Group', 'COVID-19 Deaths', 'Total Deaths',
       'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths']]
final_data_df
```

```python
us_df = final_data_df.loc[final_data_df["State"] == "United States", :]
us_df
```

```python
us_all_sex_df = us_df.loc[us_df["Sex"] == "All Sexes",:]
us_all_sex_df
```

```python
us_all_sex_by_total_df = us_all_sex_df.loc[us_all_sex_df["Group"] == "By Total",:]
us_all_sex_by_total_df
```

```python
diseases = ['COVID-19 Deaths','Pneumonia Deaths', 'Pneumonia and \nCOVID-19 Deaths']

all_total = us_all_sex_by_total_df.loc[0,['COVID-19 Deaths', 'Pneumonia Deaths', 
                                  'Pneumonia and COVID-19 Deaths']]
all_total
```

```python
plt.bar(diseases, all_total, color='purple',alpha=.3)
plt.xticks(rotation = 70, ha = 'right', rotation_mode = 'anchor')
plt.title('Full Term Total Deaths by Disease')
plt.ylabel("Volume of Death's")
plt.savefig("4Year_Death_By_Disease.png", bbox_inches = 'tight')
plt.show()
```

```python
drop_total_age_df = us_all_sex_by_total_df.loc[1:16,:]
drop_total_age_df
```

```python
test_df = drop_total_age_df.loc[[1,3,4,5,7,9,11,13,14,15,16],:]
test_df
```

```python
young_group_df = test_df.loc[1:7,:]
young_group_df
```

```python
young_ages = ['Under 1 year', '1-4 years', '5-14 years',
              '15-24 years', '25-34 years']

us_AllSex_Total_CovidDeaths_younger = young_group_df['COVID-19 Deaths']
us_AllSex_Total_PneumoniaDeaths_younger = young_group_df['Pneumonia Deaths']

x_axis = np.arange(len(young_ages))

plt.figure(figsize=(15,5))
plt.bar(x_axis, us_AllSex_Total_CovidDeaths_younger, .3, color='green', alpha=.5, label = 'Covid Deaths')
plt.bar(x_axis+.3, us_AllSex_Total_PneumoniaDeaths_younger, .3, label = 'Pneumonia Deaths')
plt.xticks(x_axis +.3/2, young_ages)
plt.xticks(rotation = 70, ha = 'right', rotation_mode = 'anchor')
plt.legend()
plt.savefig("Covid_Pneumonia_Young_Deaths.png", bbox_inches = 'tight')
plt.show()
```

```python
old_group_df = test_df.loc[9:16,:]
old_group_df
```

```python
old_ages = ['35-44 years', 
              '45-54 years','55-64 years', '65-74 years', 
              '75-84 years', '85 years and over']

us_AllSex_Total_CovidDeaths_older = old_group_df['COVID-19 Deaths']
us_AllSex_Total_PneumoniaDeaths_older = old_group_df['Pneumonia Deaths']

x_axis = np.arange(len(old_ages))

plt.figure(figsize=(15,5))
plt.bar(x_axis, us_AllSex_Total_CovidDeaths_older, .3, color = 'green', alpha = .5, label = 'Covid Deaths')
plt.bar(x_axis+.3, us_AllSex_Total_PneumoniaDeaths_older, .3, label = 'Pneumonia Deaths')
plt.xticks(x_axis +.3/2, old_ages)
plt.xticks(rotation = 70, ha = 'right', rotation_mode = 'anchor')
plt.legend()
plt.savefig("Covid_Pneumonia_Older_Deaths.png", bbox_inches = 'tight')
plt.show()
```
