```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
covid_data_csv = Path("Cleaned_Covid_Data.csv")
covid_data_df = pd.read_csv(covid_data_csv)
```

```python
covid_data_df
```

```python
#Trimming down the COVID Dataframe
columns = ["Start Date","End Date","Group","Year","State","Sex","Age Group","COVID-19 Deaths"]
covid_data_all_sexes=covid_data_df.loc[covid_data_df["Sex"]=="All Sexes",columns]
covid_data_all_ages=covid_data_all_sexes.loc[covid_data_all_sexes["Age Group"]=="All Ages",columns]

#Making Dataframes with data from New Jersey, New York, Pennsylvania, California, Texas, Florida, and Illinois
covid_nj_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="New Jersey",columns]
covid_ny_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="New York",columns]
covid_pa_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="Pennsylvania",columns]
covid_ca_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="California",columns]
covid_tx_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="Texas",columns]
covid_fl_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="Florida",columns]
covid_wy_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="Wyoming",columns]
covid_ma_df=covid_data_all_ages.loc[covid_data_all_ages["State"]=="Massachusetts",columns]

covid_nj_df
```

```python
#State population estimates in 2020
nj_2020_pop=9272392
ny_2020_pop=20104710
pa_2020_pop=12995477
ca_2020_pop=39503200
tx_2020_pop=29234361
fl_2020_pop=21591299
wy_2020_pop=577664
ma_2020_pop=6997713
```

```python
#Making Dataframes using all time data
covid_nj_alltime=covid_nj_df.loc[covid_nj_df["Group"]=="By Total",columns]
covid_ny_alltime=covid_ny_df.loc[covid_ny_df["Group"]=="By Total",columns]
covid_pa_alltime=covid_pa_df.loc[covid_pa_df["Group"]=="By Total",columns]
covid_ca_alltime=covid_ca_df.loc[covid_ca_df["Group"]=="By Total",columns]
covid_tx_alltime=covid_tx_df.loc[covid_tx_df["Group"]=="By Total",columns]
covid_fl_alltime=covid_fl_df.loc[covid_fl_df["Group"]=="By Total",columns]
covid_wy_alltime=covid_wy_df.loc[covid_wy_df["Group"]=="By Total",columns]
covid_ma_alltime=covid_ma_df.loc[covid_ma_df["Group"]=="By Total",columns]

#Bar graph showing all time data per state
states=[(covid_nj_alltime.iloc[0,4]),(covid_ny_alltime.iloc[0,4]),(covid_pa_alltime.iloc[0,4]),(covid_ca_alltime.iloc[0,4]),
              (covid_tx_alltime.iloc[0,4]),(covid_fl_alltime.iloc[0,4]),(covid_wy_alltime.iloc[0,4]),(covid_ma_alltime.iloc[0,4])]

covid_alltime_deaths=[(covid_nj_alltime.iloc[0,7]),(covid_ny_alltime.iloc[0,7]),(covid_pa_alltime.iloc[0,7]),(covid_ca_alltime.iloc[0,7]),
              (covid_tx_alltime.iloc[0,7]),(covid_fl_alltime.iloc[0,7]),(covid_wy_alltime.iloc[0,7]),(covid_ma_alltime.iloc[0,7])]

covid_alltime_per_capita=[(covid_nj_alltime.iloc[0,7])/nj_2020_pop,(covid_ny_alltime.iloc[0,7])/ny_2020_pop,
              (covid_pa_alltime.iloc[0,7])/pa_2020_pop,(covid_ca_alltime.iloc[0,7])/ca_2020_pop,
              (covid_tx_alltime.iloc[0,7])/tx_2020_pop,(covid_fl_alltime.iloc[0,7])/fl_2020_pop,
              (covid_wy_alltime.iloc[0,7])/wy_2020_pop,(covid_ma_alltime.iloc[0,7])/ma_2020_pop]

plt.bar(states,covid_alltime_deaths)
plt.xticks(rotation="vertical")
plt.title("Total COVID Deaths 2020-2023")
plt.ylabel("Total Deaths")
plt.show()
plt.bar(states,covid_alltime_per_capita,color="g")
plt.xticks(rotation="vertical")
plt.title("COVID Deaths per Capita 2020-2023")
plt.ylabel("Deaths per Capita")
plt.show()
```

```python
#Making Dataframes using yearly data
covid_nj_year=covid_nj_df.loc[covid_nj_df["Group"]=="By Year",columns]
covid_ny_year=covid_ny_df.loc[covid_ny_df["Group"]=="By Year",columns]
covid_pa_year=covid_pa_df.loc[covid_pa_df["Group"]=="By Year",columns]
covid_ca_year=covid_ca_df.loc[covid_ca_df["Group"]=="By Year",columns]
covid_tx_year=covid_tx_df.loc[covid_tx_df["Group"]=="By Year",columns]
covid_fl_year=covid_fl_df.loc[covid_fl_df["Group"]=="By Year",columns]
covid_wy_year=covid_wy_df.loc[covid_wy_df["Group"]=="By Year",columns]
covid_ma_year=covid_ma_df.loc[covid_ma_df["Group"]=="By Year",columns]

#Calculating deaths per capita per state Dataframe

covid_nj_year_percapita=[]
covid_ny_year_percapita=[]
covid_pa_year_percapita=[]
covid_ca_year_percapita=[]
covid_tx_year_percapita=[]
covid_fl_year_percapita=[]
covid_wy_year_percapita=[]
covid_ma_year_percapita=[]

for index,row in covid_nj_year.iterrows():
    covid_nj_year_percapita.append(row["COVID-19 Deaths"]/nj_2020_pop)

for index,row in covid_ny_year.iterrows():
    covid_ny_year_percapita.append(row["COVID-19 Deaths"]/ny_2020_pop)

for index,row in covid_pa_year.iterrows():
    covid_pa_year_percapita.append(row["COVID-19 Deaths"]/pa_2020_pop)

for index,row in covid_ca_year.iterrows():
    covid_ca_year_percapita.append(row["COVID-19 Deaths"]/ca_2020_pop)
    
for index,row in covid_tx_year.iterrows():
    covid_tx_year_percapita.append(row["COVID-19 Deaths"]/tx_2020_pop)

for index,row in covid_fl_year.iterrows():
    covid_fl_year_percapita.append(row["COVID-19 Deaths"]/fl_2020_pop)

for index,row in covid_wy_year.iterrows():
    covid_wy_year_percapita.append(row["COVID-19 Deaths"]/wy_2020_pop)

for index,row in covid_ma_year.iterrows():
    covid_ma_year_percapita.append(row["COVID-19 Deaths"]/ma_2020_pop)
```

```python
#Constructing the line graph
years=[]
for index,row in covid_nj_year.iterrows():
    years.append(int(row["Year"]))

tick_locations=[value for value in years]

nj_line,=plt.plot(years,covid_nj_year_percapita,label="New Jersey")
ny_line,=plt.plot(years,covid_ny_year_percapita,label="New York")
pa_line,=plt.plot(years,covid_pa_year_percapita,label="Pennsylvania")
ca_line,=plt.plot(years,covid_ca_year_percapita,label="California")
tx_line,=plt.plot(years,covid_tx_year_percapita,label="Texas")
fl_line,=plt.plot(years,covid_fl_year_percapita,label="Florida")
wy_line,=plt.plot(years,covid_wy_year_percapita,label="Wyoming")
ma_line,=plt.plot(years,covid_ma_year_percapita,label="Massachusetts")
plt.xticks(tick_locations,years)
plt.title("COVID Deaths per Capita 2020-2023")
plt.ylabel("Deaths per Capita")
plt.legend(handles=[nj_line,ny_line,pa_line,ca_line,tx_line,fl_line,wy_line,ma_line])
plt.rcParams["figure.figsize"]=(10,8)
plt.grid()
plt.show()
```
